```python
import torch 
from torch import nn 
```

```python
x = torch.randn(1, 3, 227, 227)
```

```python
relu = nn.ReLU()
```

```python
maxpool = nn.MaxPool2d(
    kernel_size = 3,
    stride = 2,
)
```

```python
dropout = nn.Dropout(p=0.5)
```

```python
conv1 = nn.Conv2d(
    in_channels = 3,
    out_channels = 96,
    kernel_size = 11,
    stride = 4,
)
```

```python
conv2 = nn.Conv2d(
    in_channels = 96,
    out_channels = 256,
    kernel_size = 5,
    stride=1,
    padding=2,
)
```

```python
conv3 = nn.Conv2d(256, 384, kernel_size=3, padding=1)
conv4 = nn.Conv2d(384, 384, kernel_size=3, padding=1)
conv5 = nn.Conv2d(384, 256, kernel_size=3, padding=1)
```

```python
x = conv1(x)
x = relu(x)
x = maxpool(x)
x = conv2(x)
x = relu(x)
x = maxpool(x)
x = conv3(x)
x = relu(x)
x = conv4(x)
x = relu(x)
x = conv5(x)
x = relu(x)
x = maxpool(x)
```

```python
x = torch.flatten(x, start_dim=1)
```

```python
print(x.shape)
```

```python
fc1 = nn.Linear(9216, 4096)
fc2 = nn.Linear(4096, 4096)
fc3 = nn.Linear(4096, 1000)
```

```python
x = fc1(x)
x = relu(x)
x = dropout(x)

x = fc2(x)
x = relu(x)
x = dropout(x)

x = fc3(x)
```
